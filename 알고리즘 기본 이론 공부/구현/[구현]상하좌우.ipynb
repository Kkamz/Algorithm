```python
n =int(input())
plans = input().split()
```

```python
x, y = 1,1
```

```python
# L R U D
dx = [0,0,-1,1]
dy = [-1,1,0,0]
move_type = ["L","R","U","D"]
```

```python
for p in plans:
    idx = move_type.index(p)
    nx = x + dx[idx]
    ny = y + dy[idx]
    if nx < 1 or ny < 1 or nx > n or ny > n:
        continue
    x, y = nx, ny
print(x,y)
```
