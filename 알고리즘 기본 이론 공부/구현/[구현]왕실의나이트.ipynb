```python
location = input()
```

```python
row = int(location[1])
col = int(ord(location[0]))-int(ord("a"))+1
```

```python
re_loc = [(-2,-1),(-2,1),(2,1),(2,-1),(1,2),(1,-2),(-1,2),(-1,-2)]
answer = 0
for l in re_loc:
    n_row = row + l[0]
    n_col = col + l[1]
    if n_row >=1 and n_row <=8 and n_col >=1 and n_col <=8:
        answer +=1
print(answer)
```
